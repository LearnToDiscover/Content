```python
from pandas import read_csv
```

```python
df = read_csv('data/everleys_data.csv')
```

```python
df.head()
```

```python
from matplotlib.pyplot import subplots
```

```python
fig, ax = subplots()
```

```python
fig, ax = subplots()

ax.boxplot(df['calcium']);
```

```python
fig, ax = subplots(figsize=(5, 6))

ax.boxplot(df['calcium']);
```

```python
fig, ax = subplots()

ax.hist(df['calcium']);
```

```python
fig, ax = subplots()

ax.hist(df['calcium'], bins=15);
```

```python
fig, ax = subplots(figsize=(8, 6), ncols=2)

ax[0].boxplot(df['calcium']);
ax[1].boxplot(df['sodium']);
```

```python
fig, ax = subplots(figsize=(10, 6), ncols=2)

ax[0].boxplot(df['calcium']);
ax[1].hist(df['calcium']);

ax[0].set_ylabel('Concentration', fontsize=16);
ax[1].set_ylabel('Count', fontsize=16);

ax[0].set_xlabel('Calcium', fontsize=16);
ax[1].set_xlabel('Calcium', fontsize=16);

fig.tight_layout()
```
